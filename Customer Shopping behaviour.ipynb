```python
df["Review Rating"] = df["Review Rating"].fillna(df.groupby("Category")["Review Rating"].transform("median"))
```

```python
df.isna().sum()
```

```python
df.columns = df.columns.str.lower().str.replace(" " , "_")
df  = df.rename(columns ={'purchase_amount_(usd)': "purchase_amount_usd"})
```

```python
df.columns
```

## feature engineering

```python
#adding age group column
labels =["Young Adult", "Adult", "Middle-age", "Senior"]
df["age_group"] = pd.qcut(df["age"],q=4, labels=labels)

                         
```

```python
df[["age","age_group"]].head()
```

```python
df.head(1)
```

```python
# purchase frequency days
df["frequency_of_purchases"].unique()
```

```python
frequency_map = {
    "Weekly": 7,
    "Bi-Weekly": 14,        
    "Fortnightly": 14,
    "Monthly": 30,
    "Every 3 Months": 90,   
    "Quarterly": 90,
    "Annually": 365
}
df["purchase_frequency_days"] = df["frequency_of_purchases"].map(frequency_map)

```

```python
df.head()
```

```python
(df["promo_code_used"] == df["discount_applied"]).all()
```

```python
df = df.drop("promo_code_used", axis=1)
```
